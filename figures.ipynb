## Error plot of trends

```python
# Any door no check
a = np.loadtxt("outputs/any_door-no_check_20_4.csv")
b = np.loadtxt("outputs/any_door-no_check_40_8.csv")
c = np.loadtxt("outputs/any_door-no_check_60_12.csv")
d = np.loadtxt("outputs/any_door-no_check_80_16.csv")

x = [4, 8, 12, 16]
y = [np.mean(a), np.mean(b), np.mean(c), np.mean(d)]
yerr = [np.std(a), np.std(b), np.std(c), np.std(d)]

plt.errorbar(x,y, yerr, fmt="o-")
plt.xlabel("# of entering/exiting passengers")
plt.ylabel("Enter/exit-time")
plt.title("Trend using any door, no ticket check")
plt.grid()
plt.show()
```
